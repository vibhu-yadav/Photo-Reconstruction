```python
for dirs, subdirs, files in os.walk('./pytorch-CycleGAN-and-pix2pix/results/pix2pix_animals/Test_latest/images/'):
    
    for file in files:
        if not '_fake_B' in file:
            continue
        
        f = file.replace('_fake_B','')
        img = cv.imread(os.path.join(dirs,file))
        cv.imwrite(f'./Dataset/Results_{attempt}/{f}',img)
```

```python
masking_info = pd.read_csv('./Dataset/masked_info.csv')
masking_info.head()

df = []

for i in range(len(masking_info)):
    
    mask = masking_info.iloc[i,:]
    _,filename, row1, col1, row2, col2 = mask[:]
    
    img = cv.imread(f'./Dataset/Results_{attempt}/{filename}')
    b,g,r = cv.split(img)
    b = cv.normalize(src = b, dst = None, alpha = 0, beta = 1, norm_type = cv.NORM_MINMAX, dtype = cv.CV_32F)
    g = cv.normalize(src = g, dst = None, alpha = 0, beta = 1, norm_type = cv.NORM_MINMAX, dtype = cv.CV_32F)
    r = cv.normalize(src = r, dst = None, alpha = 0, beta = 1, norm_type = cv.NORM_MINMAX, dtype = cv.CV_32F)
    
    print(filename, img.shape)
    print(b.shape, g.shape, r.shape)
    
    for i in range(row1,row1+75):
        for j in range(col1,col1+75):
            entry_name =  filename+"_"+'box1'+'_'+str(i)+'_'+str(j)+'_'
            df.append((entry_name+'0',b[i,j]))
            df.append((entry_name+'1',g[i,j]))
            df.append((entry_name+'2',r[i,j]))
            
    for i in range(row2,row2+75):
        for j in range(col2,col2+75):
            entry_name =  filename+"_"+'box2'+'_'+str(i)+'_'+str(j)+'_'
            df.append((entry_name+'0',b[i,j]))
            df.append((entry_name+'1',g[i,j]))
            df.append((entry_name+'2',r[i,j]))    
            
df = pd.DataFrame(df, columns = ['filename_box_pixel','Value'])
    
```
